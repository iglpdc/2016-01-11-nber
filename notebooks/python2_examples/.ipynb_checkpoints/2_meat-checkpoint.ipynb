```python
meat_ts = meat.set_index(['date'])
```

```python
meat_ts.tail()
```

A proper time series object

```python
yearly_totals = meat_ts.groupby(meat_ts.index.year).sum()
```

```python
yearly_totals.tail()
```

What if we want to examine the series by decade?  

Timestamps do not have a .decade attribute.

Let's define a function to group by decade

```python
def floor_decade(date_value):
    "Takes a date. Returns the decade."
    return (date_value.year // 10) * 10
```

```python
print floor_decade(pd.to_datetime('2015-01-08'))
print floor_decade(pd.to_datetime('1974-01-08'))
print floor_decade(pd.to_datetime('1944-01-08'))
```

```python
assert floor_decade(pd.to_datetime('2015-01-08')) == 2010, "Check floor_decade"
```

Now, we can group by our created function, "`floor_decade()`"

```python
decade_totals = meat_ts.groupby(floor_decade).sum()
```

```python
decade_totals
```

```python
decade_totals.plot(kind='bar')
```

```python
decade_totals.plot(kind='bar', stacked=True)
```

```python
meat_ts.plot()
```

```python
focus = 'beef'
meat_ts.beef.plot(style='k--', alpha=0.5, title="Beef Consumption and Moving Averages", figsize=(10,8))
pd.rolling_mean(meat_ts[focus], 60).plot(color='Blue', alpha=0.5, linewidth=3)
pd.ewma(meat_ts.ix[:,focus], 60).plot(color='Green', alpha=0.5, linewidth=2)
```

# Distributions of Changes

> Yearly totals

```python
yearly_totals.plot()
```

```python
yearly_changes = yearly_totals.diff()
```

```python
yearly_changes.plot()
```

```python
yearly_changes.head()
```

```python
yearly_changes.tail()
```

```python
yc = yearly_changes.ix[1945:2014,]
```

```python
yc.hist(bins=31, figsize=(10,5))
```

```python
meat_types = list(yc.columns)
meat_types
```

```python
for m in meat_types:
    plt.figure()
    to_plot = yc[m]
    to_plot.hist(bins=31, normed=True)
    to_plot.plot(kind='kde', title=m, linewidth=3, alpha=0.6)
    plt.show()
    
```

```python
def hist_and_kde(s2p):
    """Plot histogram and kde overlay"""
    import matplotlib.pyplot as plt
    plt.figure()
    s2p.hist(bins=31, normed=True, color="Blue")
    s2p.plot(kind='kde', linewidth=3, title=s2p.name, alpha=0.6)
    plt.show()
```

```python
for m in meat_types:
    hist_and_kde(yc[m])
```

```python
normed_yc = yc/yc.std()
```

```python
normed_yc.plot()
```

```python
normed_yc.hist(bins=31)
```

```python
normed_yc.plot(kind='kde', figsize=(10,6))
```

```python
for meat in normed_yc.columns:
    hist_and_kde(normed_yc[meat])
```
